B) Describe the relationship between a teams first round pick slot in one year with their first round slot in the subsequent year

```python
#create df with 1st round picks

fr = players[players['overall_pick'] <=30]
```

```python
fr.info()
```

```python
fr.describe()
```

```python
fr.head()
```

```python
# Matrix form for correlation data 
fr.corr()
```

```python
sns.heatmap(fr.corr())
```

```python
fr.pivot_table(values = 'rank',index= 'year',columns = 'team')
```

```python
pvfr = fr.pivot_table(values = 'rank',index= 'team',columns = 'year')
sns.heatmap(pvfr)
```

```python
pvfr.isna()
```

```python
pvfr1 = pvfr.interpolate()
```

```python
pvfr1 = pvfr.interpolate()
sns.heatmap(pvfr1, cmap='magma')
```
